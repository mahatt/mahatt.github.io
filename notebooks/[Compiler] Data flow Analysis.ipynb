### Dataflow Analysis

* static computation of infor about flow of data for each program point in the program 
* Info => a safe approximation of the desired properties of the run time behaviour  during each possible execution of that program point on all possible inputs

* semantic validity  e.g. uninitialized variables and flow of undef values
* Understanding the behaviour e.g. debugging or testing
* transformation with reasoning

* unconditional constraint resolution systems based on equalities =>  Data Flow Equations
* Static or Dynamic in case of JIT

* Scope
    + Local DFA  => in Basic Block 
    + Global DFA => in Function Across Basic Blocks
    + Interprocedural => across functions
    
* Data flow representations can be done with
    + Abstract Syntax Tree (AST)  (dont have control flow explicit)
    
    + Directed Acyclic Graph  (DAGS)
    + Control Flow Graphs (CFG) 
    
    + Program Flow Graph (PGF) (interprocedural)
    + MultiGraph (CG)  (interprocedural)
    + Program Dependence Graph (PDG) (interprocedural)
    
    + Single Static Assignment (Global and Local)
